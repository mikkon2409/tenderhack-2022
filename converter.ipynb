```python
charact = cte['Название СТЕ']
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
# lemmatizer = nltk.stem.WordNetLemmatizer()
# def lemmatize_text(text):
#     return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]
# df = pd.DataFrame(list(charact), columns=['text'])
# df['text_lemmatized'] = df.text.apply(lemmatize_text)

# df
import re

reg = re.compile('[^а-яА-Я ]')
morph = pymorphy2.MorphAnalyzer()

arr = []

count = 0
for sentence in charact:
    count += 1
    sentence = sentence.split()
    for word in sentence:
        p = morph.parse(word)[0]
        if p.tag.POS == "ADJF":
                m = p.tag.POS[:-1]
        else:
            m = p.tag.POS
        arr.append(reg.sub('', p.normal_form) + f'_{m}')
                
```

```python
charact
```

```python
def func(x):
    try:
        return json.loads(x)
    except:
        if pd.isna(x):
            return []
        else:
            if len(x) == 0:
                return []
            else:
                return json.loads(''.join([x, '"}]']))

charact = charact.apply(func)
```

```python
fields = set()
for item in charact.items():
    for entity in item[1]:
        fields.update(entity.keys())
fields
```
